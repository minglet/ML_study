```python
import pandas as pd
import numpy as numpy

movies = pd.read_csv('./ml-latest-small/movies.csv')
ratings = pd.read_csv('./ml-latest-small/ratings.csv')

print(movies.shape)
print(ratings.shape)
```

```python
movies.head()
```

```python
ratings.head()
```

```python
# 로우레벨 사용자 평점 데이터를 사용자-아이템 평점 행렬로 변환
ratings = ratings[['userId', 'movieId', 'rating']]
ratings_matrix = pd.pivot_table(ratings, 'rating', index='userId', columns='movieId')
ratings_matrix.head()
```

```python
# title 컬럼을 얻기 이해 movies 와 조인 수행
ratings_movie = ratings.merge(movies, on='movieId')

ratings_matrix = ratings_movie.pivot_table('rating', index='userId', columns='title')

# NaN 값을 모두 0 으로 변환
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix.head()
```
