```python
# 원본 dataFrame 유지, copy하기
credit_df = credit_data.copy()
# Time 칼럼 삭제
credit_df.drop('Time', axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

def get_train_test_split(df):
    X = df.iloc[:,:-1]
    y = df.iloc[:, -1]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
    
    return X_train, X_test, y_train, y_test
```

```python
# 굉장한 데이터 불균형이 있음
X_train, X_test, y_train, y_test = get_train_test_split(credit_df)
print('train data label 값')
print(y_train.value_counts())
print('test data label 값')
print(y_test.value_counts())

print('train data label 값 비율')
print(y_train.value_counts()/y_train.shape[0]*100)
print('test data label 값 비율')
print(y_test.value_counts()/y_test.shape[0]*100)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler

# StandardScaler는 Amount(labeled)에 대해서만 진행 / log 변환은 안했을 경우
scaler = StandardScaler()
amount_n = scaler.fit_transform(credit_df['Amount'].values.reshape(-1, 1))
credit_df.insert(0, 'AmountScaled', amount_n)

plt.figure(figsize=(8, 4))
plt.xticks(range(0, 30000, 1000), rotation=45)
sns.histplot(credit_df['AmountScaled'], bins=50, kde=True)
plt.show()
```

```python
# log 변환과 StandardScaler 둘다 진행하기
def preprocessing(df):
    df_copy = df.copy()
    # log 변환
    amount_log = np.log1p(df_copy['Amount'])
    df_copy.insert(0, 'Amount_log', amount_log)
    # StandardScaler 변환
    scaler = StandardScaler()
    amount_scaled = scaler.fit_transform(df_copy['Amount_log'].values.reshape(-1, 1))
    df_copy.insert(0, 'Amount_scaled', amount_scaled)
    df_copy.drop(['Time', 'Amount', 'Amount_log'], axis=1, inplace=True)
    
    return df_copy
```

```python
credit_pre_df = preprocessing(credit_data)
credit_pre_df.head()
```

```python
# 모델 평가 함수 생성
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score

def get_clf_eval(y_test, pred=None, pred_proba=None):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    
    # ROC-AUC
    roc_auc = roc_auc_score(y_test, pred_proba)
    print('오차 행렬')
    print(confusion)
    
    print('accuracy: {0:.3f}, precision: {1:.3f}, recall: {2:.3f}, f1 score: {3:.3f}, AUC : {4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
# 모델 평가 및 예측 함수 생성
def model_train_eval(model, X_train=None, X_test=None, y_train=None, y_test=None):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    pred_proba = model.predict_proba(X_test)[:, 1]
    get_clf_eval(y_test, pred, pred_proba)
```

```python
# train, test set 나누기
X_train, X_test, y_train, y_test = get_train_test_split(credit_pre_df)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier

lr_clf = LogisticRegression(max_iter=1000)
print(model_train_eval(lr_clf, X_train, X_test, y_train, y_test))

# LGBM
lgb_clf = LGBMClassifier(n_estimators=400, num_leaves=64, n_jobs=-1, boost_from_average=False)
print(model_train_eval(lgb_clf, X_train, X_test, y_train, y_test))
```

```python
import numpy as np

def get_outlier(df=None, column=None, weight=1.5):
    # fraud에 해당하는 column 데이터만 추출, 1/4 분위와 3/4 분위 지점을 np.percentile로 구하기
    fraud = df[df['Class']==1][column]
    quantile_25 = np.percentile(fraud.values, 25) # 백분위 25에 해당하는 부분의 값이 들어감
    quantile_75 = np.percentile(fraud.values, 75) # 백분위 75에 해당하는 부분의 값이 들어감
    # IQR을 구하고, IQR에 1.5를 곱하여 최대값과 최소값 지점 구함
    iqr = quantile_75 - quantile_25
    iqr_weight = iqr*weight
    lowest_val = quantile_25 - iqr_weight
    hightest_val = quantile_75 + iqr_weight
    
    # 최대값 보다 크거나, 최소값보다 작은 값을 아웃라이어로 설정하고 dataframe index 반환
    outlier_index = fraud[(fraud < lowest_val) | (fraud > hightest_val)].index
    return outlier_index
```

```python
fraud = credit_pre_df[credit_pre_df['Class']==1]['V14']
# column에 있는 모든 value값을 가지고 퍼센테이지로 나누기
np.percentile(fraud.values, 25) # precentile(values, q=퍼센트)
```

```python
# 이상치 데이터 인덱스 뽑기
outlier_index = get_outlier(credit_pre_df, 'V14', weight=1.5)
print(outlier_index)
```

```python
# preprocessing 재정의
# log1p 적용
# outlier 제거
# 필요없는 칼럼 제거 ('Amount', 'Time')
# array.insert(삽입할 index,'column name', item)
import numpy as np

def preprocessing(df):
    df_copy = df.copy()
    # log1p 적용
    scaled_amount = np.log1p(df_copy['Amount'])
    df_copy.insert(0, 'Amount_scaled', scaled_amount)
    df_copy.drop(['Amount', 'Time'], axis=1, inplace=True)
    # outlier 제거
    outlier_index = get_outlier(df_copy, 'V14', weight=1.5)
    df_copy.drop(outlier_index, axis=0, inplace=True)
    return df_copy
    
```

```python
import pandas as pd

df = pd.read_csv('../data/creditcard.csv')
pre_df = preprocessing(df)
pre_df.head(3)
```

```python
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier

# train test set 나누고 학습 / 예측
X_train, X_test, y_train, y_test = get_train_test_split(pre_df)

# Logistic Regression
lr_clf = LogisticRegression(max_iter=1000)
print(model_train_eval(lr_clf, X_train, X_test, y_train, y_test))

# LGBM
lgb_clf = LGBMClassifier(n_estimators=400, num_leaves=64, n_jobs=-1, boost_from_average=False)
print(model_train_eval(lgb_clf, X_train, X_test, y_train, y_test))
```

### SMOTE 오버샘플링

```python
print(y_train.value_counts())
```

```python
from imblearn.over_sampling import SMOTE
# over sampling의 중요한 점! 테스트셋은 건들면 안된다!

smote = SMOTE(random_state=0)
X_train_over, y_train_over = smote.fit_resample(X_train, y_train)
print('SMOTE 적용 전 학습용 피처/레이블 데이터 셋: ', X_train.shape, y_train.shape)
print('SMOTE 적용 후 학습용 피처/레이블 데이터 셋: ', X_train_over.shape, y_train_over.shape)
print('SMOTE 적용 후 레이블 값 분포: \n', y_train_over.value_counts())
```

```python
# Oversampling 후 학습 및 예측
# Logistic Regression
lr_clf = LogisticRegression(max_iter=1000)
print(model_train_eval(lr_clf, X_train_over, X_test, y_train_over, y_test))

# LGBM
lgb_clf = LGBMClassifier(n_estimators=400, num_leaves=64, n_jobs=-1, boost_from_average=False)
print(model_train_eval(lgb_clf, X_train_over, X_test, y_train_over, y_test))
```
